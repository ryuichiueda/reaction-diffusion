```python
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure()
ax = fig.add_subplot(111)

def update(frame, d, ax):
    d[0][0] += 1
    ax.pcolor(d, cmap=plt.cm.jet)
    
d = [[1,1],[2,3]]
anim = FuncAnimation(fig, update, frames=range(8), fargs=(d, ax), interval=1000)
plt.show()
```
